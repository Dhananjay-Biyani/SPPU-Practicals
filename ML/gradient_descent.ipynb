```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def cost(x):
    return (x+3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
initial_x = 2
learning_rate = 0.1
iterations = 25
```

```python
x_values = []
y_values = []
x = initial_x

for i in range(iterations):
    x_values.append(x)
    y_values.append(cost(x))
    gradient_value = derivative(x)
    x = x - learning_rate*gradient_value

    print(f" Iteration {i+1}: x={x} ,cost ={cost(x)} ")
print(f"optimal x:{x}")
```

```python
plt.plot(x_values , y_values,'*-')
plt.xlabel("X")
plt.ylabel("Y")
plt.show
```
